```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Kategorik olarak işlem yapılacak sütun adını belirle
categorical_column = ['Protocol']

# OneHotEncoder'ı başlat ve sparse_output=False ile çıktı formatını ayarla
ohe = OneHotEncoder(sparse_output=False)

# Kategorik sütunu dönüştür
protocol_encoded = ohe.fit_transform(csv[categorical_column])

# Dönüşen veriden yeni bir DataFrame oluştur
protocol_encoded_df = pd.DataFrame(protocol_encoded, columns=ohe.get_feature_names_out(categorical_column))

# Eski 'Protocol' sütununu orijinal DataFrame'den kaldır
csv = csv.drop(columns=categorical_column)

# Yeni one-hot encoded sütunları orijinal DataFrame'e ekle
csv = pd.concat([csv, protocol_encoded_df], axis=1)

print("Veri seti One-Hot Encoding işleminden sonraki durumu:")
print(csv.head())
```

```python
# Tüm DataFrame'deki NaN değerlerini kontrol et ve toplamını göster
print("NaN değerlerinin toplamı (temizlemeden önce):")
print(csv.isnull().sum().sum())

# Sonsuz (infinity) değerleri NaN ile değiştir
csv.replace([np.inf, -np.inf], np.nan, inplace=True)

# Sayısal ve metinsel sütunlardaki NaN değerlerini doldurma
for column in csv.columns:
    if csv[column].dtype in ['int64', 'float64']:
        # Sayısal sütunları medyan ile doldur
        median_value = csv[column].median()
        csv[column] = csv[column].fillna(median_value)
    elif csv[column].dtype == 'object':
        # Metinsel (object) sütunları mod (en sık tekrar eden değer) ile doldur
        mode_value = csv[column].mode()[0]
        csv[column] = csv[column].fillna(mode_value)

# Temizleme sonrası kalan NaN değerlerini kontrol et
print("\nNaN değerlerinin toplamı (temizledikten sonra):")
print(csv.isnull().sum().sum())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Veriyi eğitim ve test setlerine ayır
def train_val_test_split(csv, rstate=42, shuffle=True, stratify=None):
    strat = csv[stratify] if stratify else None
    train_set, test_set = train_test_split(
        csv, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)

train_set, val_set, test_set = train_val_test_split(csv, stratify='Label')
```

```python
# Separate features (X) and labels (y) for each set
X_train = train_set.drop('Label', axis=1)
y_train = train_set['Label']

X_val = val_set.drop('Label', axis=1)
y_val = val_set['Label']

X_test = test_set.drop('Label', axis=1)
y_test = test_set['Label']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the validation and test sets using the same scaler
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train, y_train)
```

```python

from sklearn.metrics import classification_report, confusion_matrix

# Modelin test verisi üzerinde tahmin yapmasını sağla
y_pred = svm_model.predict(X_test)

# Sınıflandırma raporunu yazdır
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Karmaşıklık matrisini (confusion matrix) oluştur ve yazdır
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    w = svm_clf.coef_[0]
    b = svm_clf.intercept_[0]

    # At the decision boundary, w0*x0 + w1*x1 + b = 0
    # => x1 = -w0/w1 * x0 - b/w1
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0]/w[1] * x0 - b/w[1]

    margin = 1/w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin

    svs = svm_clf.support_vectors_
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#FFAAAA')
    plt.plot(x0, decision_boundary, "k-", linewidth=2)
    plt.plot(x0, gutter_up, "k--", linewidth=2)
    plt.plot(x0, gutter_down, "k--", linewidth=2)
```
